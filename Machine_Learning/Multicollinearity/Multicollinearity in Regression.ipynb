<h2><div style="font-family: Trebuchet MS; background-color: red; color: #FFFFFF; padding: 12px; line-height: 1.5;">üìä What is Multicollinearity? ü§î</div> 

Multicollinearity refers to a situation in regression analysis where two or more independent variables are highly correlated. When this happens, it becomes challenging to pinpoint the individual effect of each variable on the dependent variable since they move together. üìà The presence of multicollinearity can lead to unstable coefficient estimates, which might fluctuate significantly upon adding or removing predictor variables. Furthermore, it obscures the importance of individual predictors.

<h3><div style="font-family: Trebuchet MS; background-color:#176BA0;; color: #FFFFFF; padding: 10px; line-height: 1.5;">1. | Why It's Important üåü üìö</div></h3>

Understanding and addressing multicollinearity is not just a theoretical concern ‚Äì it has practical implications in various fields and applications. Here's why it's vital
    
    
<h10><div style="font-family: Trebuchet MS; background-color:orange; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 | Reliable Coefficient Estimates üìà:</div></h10>

Multicollinearity can cause regression coefficients to be poorly estimated. This means that the influence of individual predictors on the outcome might not be correctly represented.

<h10><div style="font-family: Trebuchet MS; background-color:orange; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 | Model Interpretability üìñ: :</div></h10>

A primary goal of regression analysis is to understand how different variables influence the outcome. When multicollinearity is present, it becomes challenging to decipher the individual impact of each predictor. This clouds our understanding and diminishes the model's utility as a tool for insight.

<h10><div style="font-family: Trebuchet MS; background-color:orange; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 | Model Generalization üåç:</div></h10>

Models plagued with multicollinearity might not generalize well to new, unseen data. This is because the unstable coefficient estimates can lead to erratic predictions when applied to different datasets.

<h10><div style="font-family: Trebuchet MS; background-color:orange; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 | Model  Increased Model Complexity üåÄ: </div></h10>

Multicollinearity can make models more complex than necessary. A model that includes redundant predictors might require more data to train effectively, be harder to interpret, and be prone to overfitting.

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 | Detecting and Handling Multicollinearity ‚úÖüëç</div></h10>

<h3><div style="font-family: Trebuchet MS; background-color:#176BA0;; color: #FFFFFF; padding: 10px; line-height: 1.5;">1. | Tackling Multicollinearity - Practical Implementation üõ†Ô∏è</div></h3>




<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 | Step 1: Check Correlations üìè</div></h10>
Examine the linear relationships between independent variables. High correlation values between variables suggest possible multicollinearity.

```python
# Import necessary functions
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
np.random.seed(42)
house_size = np.random.randint(1800, 2600, 100)
num_bedrooms = (house_size / 500).astype(int)
house_age = np.random.randint(1, 10, 100)
house_price = 5000 + 100 * house_size + -5000 * num_bedrooms + -3000 * house_age + np.random.normal(0, 10000, 100)

data = pd.DataFrame({
    'House Size (sq.ft)': house_size,
    'Number of Bedrooms': num_bedrooms,
    'House Age (yrs)': house_age,
    'House Price ($)': house_price
})

# Calculate and visualize correlations
correlation_matrix = data.corr()
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

From the correlation matrix heatmap, we can observe:
    
<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Size vs. Number of Bedrooms üìêüõå: </div></h10>

A very high correlation (~0.88). This indicates that as the house size increases, the number of bedrooms tends to increase as well. üìà

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Size vs. House Price üìêüí≤: </div></h10>

A very high correlation (~0.82) suggests that larger houses tend to have higher prices. üè†üí∞

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |Number of Bedrooms vs. House Age üõåüí≤:  </div></h10>

A weak negative correlation (-0.06), which suggests that the number of bedrooms has a slight negative influence on the House Age.

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Age vs. House Price ‚è≥üí≤: </div></h10>

A negative correlation (-0.33) indicates that older houses tend to have lower prices. üèöÔ∏èüìâ


    
<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |Step 2: Calculate VIF for Features üìê</div></h10>
Next, we'll calculate the Variance Inflation Factor (VIF) for each feature to assess the severity of multicollinearity.

```python
# Import necessary functions
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF
features = data.drop("House Price ($)", axis=1)
vif_data = pd.DataFrame()
vif_data["VIF"] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
vif_data["Feature"] = features.columns
vif_data
```

Here's what the VIF values suggest:

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Size (sq.ft) üìê: </div></h10>
    
A VIF of approximately 171.55, which is significantly above the common threshold of 5-10, indicating that this variable is highly multicollinear. üö´üî¥

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |Number of Bedrooms üõå: </div></h10>

A VIF of approximately 157.60, again indicating a high degree of multicollinearity. üö´üî¥
    
<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Age (yrs) ‚è≥: </div></h10>

A VIF of approximately 4.28, suggesting that this variable is not as multicollinear as the other two. ‚úÖüü¢

The high VIF values for 'House Size' and 'Number of Bedrooms' confirm our earlier observation from the correlation matrix: these two variables are highly correlated and might be causing multicollinearity issues in our regression model. üìä‚ùó


<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |Step 3: Regularize Your Model üåä</div></h10>    
Let's implement Ridge Regression, which is a regularization technique designed to address multicollinearity, and visualize the coefficient values for each feature.

```python
from sklearn.linear_model import Ridge

# Split data into features and target
X = data.drop("House Price ($)", axis=1)
y = data["House Price ($)"]

# Implement Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)

# Visualize coefficients
coefs = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': ridge.coef_
})

plt.figure(figsize=(10, 5))
sns.barplot(data=coefs, x='Feature', y='Coefficient', palette='viridis')
plt.title('Ridge Regression Coefficients')
plt.ylabel('Coefficient Value')
plt.show()
```

The bar chart above visualizes the coefficient values from the Ridge Regression model üìà:


<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Size (sq.ft) üìê: </div></h10>
The coefficient is positive, indicating that as the house size increases, the house price tends to increase. üè†üíπ

<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |Number of Bedrooms üõå: </div></h10>

The coefficient is negative. This means that, holding other factors constant, an increase in the number of bedrooms would decrease the house price. This might seem counterintuitive, but remember that this variable is highly correlated with 'House Size'. The negative coefficient could be Ridge Regression's way of balancing the effects of 'House Size' and 'Number of Bedrooms' due to multicollinearity. ü§π‚Äç‚ôÇÔ∏èüîÄ
    
<h10><div style="font-family: Trebuchet MS; background-color:Black; color: #FFFFFF; padding: 1px; line-height: 1.;">1.1 |House Age (yrs) ‚è≥: </div></h10>
